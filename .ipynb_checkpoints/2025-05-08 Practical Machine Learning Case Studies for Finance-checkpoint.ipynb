For the given two datasets, construct a combined dataframe with relevant features including subjectivity

The **Valence Aware Dictionary and Sentiment Reasoner (VADER)** is a lexicon and rule-based sentiment analysis tool designed to evaluate the sentiment of text, particularly in social media contexts. It is effective at detecting sentiment intensity and works well on short, informal texts like tweets, comments, and reviews.  
**Linear Discriminant Analysis (LDA)** is a statistical technique used for dimensionality reduction and classification in machine learning. It aims to find a linear combination of features that best separates multiple classes.  
Pros:
- Works well for linearly separable data.
- Reduces dimensionality while preserving class information.
- Useful when feature correlation exists.

Cons:
- Assumes normal distribution of data (not always true).
- Performs poorly on nonlinear data.
- Not as flexible as SVM or tree-based classifiers.

LDA could be useful for dimensionality reduction before applying more complex models.

### Import the libraries. ##

```python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# Import sklearn & classifier
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import classification_report, RocCurveDisplay
# Import text and sentiment libraries
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

### Load the data. ###

```python
# Read DJIA news
df1 = pd.read_csv('data/djia_news.csv')
# Read DJIA prices, tab separated
df2 = pd.read_csv('data/djia_prices.csv', sep='\t')
```

### Process the data. ###

```python
# Merge the two dataframe
merge = df1.merge(df2, how='inner', on='Date')
merge.set_index('Date', inplace=True)
```

```python
merge.head()
```

```python
# Combine headlines
headlines = []
for row in range(0, len(merge.index)):
    headlines.append(' '.join(str(x) for x in merge.iloc[row, 1:26]))
```

```python
headlines[0]
```

```python
# Clean the data
_headlines = []
# Remove b', b" and \'
for i in range(0, len(headlines)):
    _headlines.append(re.sub("b[(')]", '', headlines[i]))
    _headlines[i] = re.sub('b[(")]', '', _headlines[i])
    _headlines[i] = re.sub("\'", '', _headlines[i])
```

```python
# Verify the cleaned headlines
_headlines[0]
```

```python
# Add clean headlines to merged dataframe
merge['Combined_news'] = _headlines
# Verify
merge['Combined_news']['2016-07-01']
```

### Subjectivity & Polarity ###
- Subjectivity refers to a personal opinion or judgement. It can be viewed as subjectivity to
being objective or as an objective scale ranging from 0 to 1. A 0 indicates objective and 1
indicates subjective.
- Polarity ranges from -1 to 1. A +1 indicates a positive statement and -1 indicates a negative
statement.

```python
# Create a function to get the subjectivity
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

# Create a function to get the polarity
def getPolarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
# Add subjectivity and polarity to merged dataframe
merge['Subjectivity'] = merge['Combined_news'].apply(getSubjectivity)
merge['Polarity'] = merge['Combined_news'].apply(getPolarity)
```

```python
merge.head()
```

### Sentiment Analysis ###
Sentiment analysis is performed using VADER Sentiment Analyzer. The compound score is computed
by summing the valence scores of each word in the lexicon, adjusted according to the rules,
and then normalized to be between -1 (most extreme negative) and +1 (most extreme positive).  
The standardized thresholds for classifying sentences as either positive, neutral, or negative is also
used in research. Typical threshold values are:
- positive sentiment: compound score >= 0.05
- neutral sentiment: (compound score > -0.05) and (compound score < 0.05)
- negative sentiment: compound score <= -0.05  

The compound score is the one most commonly used metric for sentiment analysis.

```python
# Create function for sentiment scores
def getSentiment(text):
    sa = SentimentIntensityAnalyzer()
    sentiment = sa.polarity_scores(text)
    return sentiment
```

```python
# Get the sentiment scores for each day
compound = []
negative = []
positive = []
neutral = []
sa = 0
for i in range(0, len(merge)):
    sa = getSentiment(merge['Combined_news'][i])
    compound.append(sa['compound'])
    negative.append(sa['neg'])
    neutral.append(sa['neu'])
    positive.append(sa['pos'])
```

```python
compound[:5]
```

```python
# Store the sentiment scores in the dataframe
merge['Compound'] = compound
merge['Negative'] = negative
merge['Neural'] = neutral
merge['Positive'] = positive
```

```python
merge.head()
```

### Fit and Predict ###

```python
df = merge[['Open', 'High', 'Low', 'Volume', 'Subjectivity', 'Polarity', 'Compound', 'Negative', 'Neural', 'Positive', 'Label']]
# Create feature set
X = np.array(df.drop(['Label'], axis=1))
# Create target
y = np.array(df['Label'])
```

```python
print(X.shape)
print(y.shape)
```

```python
# Split the train and test dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=23, shuffle=False)
```

```python
X_train[:5]
```

```python
y_train[:5]
```

```python
X_train
```

```python
# Create and train the model
model = LinearDiscriminantAnalysis().fit(X_train, y_train)
# Prediction
y_pred = model.predict(X_test)
```

```python
# Output prediction scoare
print(f'Train Accuracy: {model.score(X_train,y_train):0.4}')
print(f'Test Accuracy: {model.score(X_test,y_test):0.4}')
```

```python
# Classification Report
print(classification_report(y_test, y_pred))
```

```python
# Display ROCCurve
disp_roc = RocCurveDisplay.from_estimator(
    model,
    X_test,
    y_test,
    name='LDA')
disp_roc.ax_.set_title('ROC Curve')
plt.plot([0,1],[0,1], linestyle='dashed')
plt.show()
```

The Linear Discriminant model is thus able to classify the labels based on the sentiments with an
accuracy of 87%.
